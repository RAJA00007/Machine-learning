Simple linear regression


```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("c:\\Users\\RAJA\\Downloads\\canada_per_capita_income.csv")
df.head()
```

```python
df.rename(columns={'per capita income (US$)': 'per_capita_income'}, inplace=True)
sns.set_style("whitegrid")
sns.scatterplot(x='year', y='per_capita_income', data=df, color = 'red')
```

```python


# Create a linear regression model
reg = linear_model.LinearRegression()

# Fit the model using the 'year' column as the feature and 'per capita income' as the target
reg.fit(df[['year']], df.per_capita_income)
```

```python
# Predicting income for a given year
reg.predict([[1971]])
```

```python
reg.coef_ 
```

```python
reg.intercept_ 
```

```python
df.rename(columns={'per capita income (US$)': 'per_capita_income'}, inplace=True)
sns.set_style("whitegrid")
sns.scatterplot(x='year', y='per_capita_income', data=df, color = 'red')
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
```
