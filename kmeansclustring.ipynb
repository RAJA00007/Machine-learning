```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("c:\\Users\\RAJA\\Downloads\\income.csv")
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
k = KMeans(n_clusters=3)
k
```

```python
y= k.fit_predict(df[['Age','Income($)']])
y
```

```python
df['cluster'] = y
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color ='green')
plt.scatter(df2.Age,df2['Income($)'],color ='red')
plt.scatter(df3.Age,df3['Income($)'],color ='black')

plt.xlabel('Age')
plt.ylabel('income($)')
plt.legend()
```

```python
sacler = MinMaxScaler()
sacler.fit(df[['Income($)']])
df['Income($)'] = sacler.transform(df[['Income($)']])
df
sacler.fit(df[['Age']])
df.Age = sacler.transform(df[['Age']])
df
```

```python
k = KMeans(n_clusters=3)
y= k.fit_predict(df[['Age','Income($)']])
y
```

```python
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color ='green')
plt.scatter(df2.Age,df2['Income($)'],color ='red')
plt.scatter(df3.Age,df3['Income($)'],color ='black')
plt.scatter(k.cluster_centers_[:,0],k.cluster_centers_[:,1],color='purple',marker='*')
plt.legend()
```

```python
k.cluster_centers_
```

```python
k_rang = range(1,10)
sse =[]
for k in k_rang:
 k=KMeans(n_clusters=k)
 k.fit(df[['Age','Income($)']])
 sse.append(k.inertia_)
    
```

```python
sse
```

```python
plt.xlabel('k')
plt.ylabel('sum of sse ')
plt.plot(k_rang,sse)
```
